```python
import numpy as np 
import pandas as pd 
data=pd.read_csv('/kaggle/input/countries-gdp-19602020/Countries GDP 1960-2020.csv')
data=data.drop("Country Code",axis=1)

# drop the null data 
data.dropna(inplace=True)
# fill the null data by the numper(130)
#data.fillna(130,inplace=True)

data

```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
lb=LabelEncoder()
cat_data = data['Country Name']
cat_data=lb.fit_transform(cat_data)
data['Country Name']=cat_data
data
```

```python
### China GDP
china = data[data["Country Name"]== 'China']
china = china.drop(['Country Name'], axis = 1)
china = china.T
china.columns = ['GDP in US$']
china_df_gdp = china.reset_index()
china_df_gdp.columns = ['Year', 'GDP in US$']
### Australia GDP
australia = data[data["Country Name"]== 'Australia']
australia = australia.drop(['Country Name'], axis = 1)
australia = australia.T
australia.columns = ['GDP in US$']
aus_df_gdp = australia.reset_index()
aus_df_gdp.columns = ['Year', 'GDP in US$']
### France GDP
france = data[data["Country Name"]== 'France']
france = france.drop(['Country Name'], axis = 1)
france = france.T
france.columns = ['GDP in US$']
france_df_gdp = france.reset_index()
france_df_gdp.columns = ['Year', 'GDP in US$']
### United Kingdom GDP
uk = data[data["Country Name"]== 'United Kingdom']
uk = uk.drop(['Country Name'], axis = 1)
uk = uk.T
uk.columns = ['GDP in US$']
uk_df_gdp = uk.reset_index()
uk_df_gdp.columns = ['Year', 'GDP in US$']
### Japan GDP
japan = data[data["Country Name"]== 'Japan']
japan = japan.drop(['Country Name'], axis = 1)
japan = japan.T
japan.columns = ['GDP in US$']
jp_df_gdp = japan.reset_index()
jp_df_gdp.columns = ['Year', 'GDP in US$']
```

```python
y_pred_train_rf=[]
def RandomForestRegressor_model(n):
    from sklearn.ensemble import RandomForestRegressor
    RandomForestRegressor_model = RandomForestRegressor(n_estimators=n, random_state=42)
    RandomForestRegressor_model.fit(x_train, y_train)
    score_train = RandomForestRegressor_model.score(x_train, y_train)
    score_test = RandomForestRegressor_model.score(x_test, y_test)
    y_pred_train_rf = RandomForestRegressor_model.predict(x_train)
    y_pred_test_rf = RandomForestRegressor_model.predict(x_test)
    print('score_train :',score_train)
    
    print('score_test :',score_test)
    return y_pred_train_rf,y_pred_test_rf
```

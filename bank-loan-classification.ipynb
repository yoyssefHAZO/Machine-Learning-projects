```python
import numpy as np 
import pandas as pd 
import cv2
import os
import tensorflow as tf
import keras 
import matplotlib.pyplot as plt
%matplotlib inline
import glob as gb
```

```python
data=pd.read_csv('/kaggle/input/bank-loan-classification/UniversalBank.csv')
data=data.drop("ID",axis=1)
data=data.drop("Online",axis=1)
data=data.drop("CreditCard",axis=1)
data=data.drop("ZIP Code",axis=1)
# drop the null data 
data.dropna(inplace=True)
# fill the null data by the numper(130)
#data.fillna(130,inplace=True)

data
```

##preprocessing 

```python
y=data['Personal Loan']
x=data.drop("Personal Loan",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

# Scaler from -1 to 1

```python
#Scaler from -1 to 1
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
def knn_model(k):
    from sklearn.neighbors import KNeighborsClassifier
    model_knn=KNeighborsClassifier(n_neighbors=k)
    model_knn.fit(x_train,y_train)
    y_predict_train=model_knn.predict(x_train)
    y_predict_test=model_knn.predict(x_test)
    from sklearn.metrics import accuracy_score
    acc_train=accuracy_score( y_train,y_predict_train)
    acc_test=accuracy_score( y_test,y_predict_test)
    print('acc of train ',acc_train)
    print('acc of test ',acc_test)
```

#  hyperparameter

```python
for i in range(100):
    print(i+1)
    knn_model(i+1)
```

```python
print('the best value of k=1')
knn_model(1)
```

```python
from sklearn.metrics import confusion_matrix

# Training the Decision Tree Classification model on the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy')
classifier.fit(x_train, y_train)
# Predicting the Test set results
y_pred = classifier.predict(x_test)
y_pred_t = classifier.predict(x_train)
# Making the Confusion Matrix
aac_train = confusion_matrix(y_train, y_pred_t)
acc_test = confusion_matrix(y_test, y_pred)
print('acc of train ',aac_train)
print('acc of test ',acc_test)
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(x_train, y_train)
y_pred_t= nb.predict(x_train)
y_pred = nb.predict(x_test)
acc_train=accuracy_score( y_train,y_pred_t)
acc_test=accuracy_score( y_test,y_pred)
print('acc of train ',acc_train)
print('acc of test ',acc_test)
```

# Random Forest Classifier

```python
def    RandomForest_model(n): 
    from sklearn.ensemble import RandomForestClassifier
    RandomForest_model = RandomForestClassifier(n_estimators=n)
    RandomForest_model.fit(x_train, y_train)
    y_pred_t = RandomForest_model.predict(x_train)
    y_pred = RandomForest_model.predict(x_test)
    acc_train=accuracy_score( y_train,y_pred_t)
    acc_test=accuracy_score( y_test,y_pred)
    print('acc of train ',acc_train)
    print('acc of test ',acc_test)
```
